# Plotting a normal distribution

```python
%matplotlib inline

import numpy as np
from scipy import stats, integrate
import matplotlib.pyplot as plt

# we wouldn't actually use seaborn
# but it makes the plots nicer.
import seaborn as sns
sns.set(color_codes=True)
```

To plot a normal distribution we will use the "norm" funcitonality from scipy.stats:

```python
from scipy.stats import norm
```

To plot the normal distribution, we'll use the "probability density function" (pdf) for the normal distribution is given by:

`norm.pdf(x)`

In simple words, the pdf(x) function answers the question:

`norm.pdf(x) = Assuming x is normaly distributed, what is the probability of getting the value in x?`

The `norm.pdf(x, loc, scale)` function can take three parameters: `x` is simply the value we're interested in, `loc` is the center of the distribution or $\mu$ and `scale` is for the standard deviation $\sigma$. 

We will make an array of `x` values, calculate the pdf for all these and store them at array `y`, and plot. To make sure everything works as expected, let us also plot another normal distribution with $\mu=1$ and $\sigma=0.75$:

```python
x = np.linspace(-5, 5, 100)
y = norm.pdf(x, loc=0, scale=1)
y2 = norm.pdf(x, loc=1, scale=0.75)

plt.plot(x, y,'b-',
         x, y2,'r-', 
         linewidth=4.0, 
         alpha=0.75)

plt.title('The normal distribution')
```

It is important to note that the normal distibution curve can be obtained using functions other than the pdf.
