# Sum of Binomially distributed RVSs

```python
import numpy as np
import scipy.stats as stats
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## The sum of two binomial distributions is a binomial distribution
Let `X` and `Y` be two random variables that are Binomially distributed, with probabilities `pX` and `pX`:

```python
n = 1000
size = 1000
pX = 0.3
pY = 0.6
X = stats.binom.rvs(n, pX, size=size, random_state=770)
Y = stats.binom.rvs(n, pY, size=size, random_state=770)
```

The sum of the two Binomial RVSs is also Binomial, as shown [here](https://math.stackexchange.com/a/1176442) and demonstrated in the figure below:

```python
sns.distplot(X, label='X')
sns.distplot(Y, label='X')
sns.distplot(X + Y, label='X+Y')
plt.legend();
```
